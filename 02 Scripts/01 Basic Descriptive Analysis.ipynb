04 Run Consistency Checks

```python
#check for mixed-type data

for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

The check returned no resuts, suggesting there is no mixed-type data.
